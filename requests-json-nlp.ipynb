The following code accesses the [lyrics.ovh](https://lyricsovh.docs.apiary.io/#reference/0/lyrics-of-a-song/search) public api, searches for the lyrics of a song, and stores it in a dictionary object.  Write the resulting json to a file (either a JSON file or a pickle file; you choose). You will read in the contents of this file for future questions so we do not need to frequently access the API.

```python
import requests
import json

result = json.loads(requests.get('https://api.lyrics.ovh/v1/They%20Might%20Be%20Giants/Birdhouse%20in%20your%20soul').text)
print(result)

import requests
import json

def get_song_lyrics(artist, title):
    url = f"https://api.lyrics.ovh/v1/{artist}/{title}"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        return None

def main():
    # Example: Search for lyrics of the song "Shape of You" by Ed Sheeran
    artist = "Ed Sheeran"
    title = "Shape of You"

    # Get the lyrics
    lyrics_data = get_song_lyrics(artist, title)

    if lyrics_data:
        # Write the JSON data to a file
        with open("lyrics_data.json", "w") as json_file:
            json.dump(lyrics_data, json_file)
            print("Lyrics data has been successfully written to lyrics_data.json")
    else:
        print("Failed to retrieve lyrics data.")

if __name__ == "__main__":
    main()
```

2. Read in the contents of your file.  Print the lyrics of the song (not the entire dictionary!) and use spaCyTextBlob to perform sentiment analysis on the lyrics.  Print the polarity score of the sentiment analysis.  Given that the range of the polarity score is `[-1.0,1.0]` which corresponds to how positive or negative the text in question is, do you think the lyrics have a more positive or negative connotaion?  Answer this question in a comment in your code cell.

```python
import json
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob

# Load spaCy model with TextBlob
nlp = spacy.load("en_core_web_sm")
nlp.add_pipe("spacytextblob")

def read_lyrics_from_file(filename):
    with open(filename, "r") as json_file:
        lyrics_data = json.load(json_file)
    return lyrics_data

def perform_sentiment_analysis(lyrics):
    doc = nlp(lyrics)
    polarity_score = doc._.polarity
    return polarity_score

def main():
    # Read the contents of the file
    lyrics_data = read_lyrics_from_file("lyrics_data.json")

    # Extract the lyrics
    lyrics = lyrics_data["lyrics"]

    # Print the lyrics of the song
    print("Lyrics of the song:")
    print(lyrics)
    print()

    # Perform sentiment analysis
    polarity_score = perform_sentiment_analysis(lyrics)
    
    # Print the polarity score of the sentiment analysis
    print("Polarity score of the sentiment analysis:", polarity_score)

    # Interpretation
    if polarity_score > 0:
        print("The lyrics have a more positive connotation.")
    elif polarity_score < 0:
        print("The lyrics have a more negative connotation.")
    else:
        print("The lyrics have a neutral connotation.")

if __name__ == "__main__":
    main()
```

3. Write a function that takes an artist, song, and filename, accesses the lyrics.ovh api to get the song lyrics, and writes the results to the specified filename.  Test this function by getting the lyrics to any four songs of your choice and storing them in different files.

```python
import requests
import json

def get_and_write_lyrics(artist, song, filename):
    url = f"https://api.lyrics.ovh/v1/{artist}/{song}"
    response = requests.get(url)
    
    if response.status_code == 200:
        lyrics_data = response.json()
        lyrics = lyrics_data.get('lyrics')
        
        if lyrics:
            with open(filename, 'w') as file:
                file.write(lyrics)
            print(f"Lyrics for '{song}' by {artist} successfully written to {filename}.")
        else:
            print(f"No lyrics found for '{song}' by {artist}.")
    else:
        print(f"Failed to retrieve lyrics for '{song}' by {artist}.")


# Test the function by getting the lyrics of four songs and storing them in different files
if __name__ == "__main__":
    get_and_write_lyrics("Ed Sheeran", "Shape of You", "shape_of_you_lyrics.txt")
    get_and_write_lyrics("Adele", "Someone Like You", "someone_like_you_lyrics.txt")
    get_and_write_lyrics("Coldplay", "Fix You", "fix_you_lyrics.txt")
    get_and_write_lyrics("Taylor Swift", "Love Story", "love_story_lyrics.txt")
```

4. Write a function that takes the name of a file that contains song lyrics, loads the file, performs sentiment analysis, and returns the polarity score.  Use this function to print the polarity scores (with the name of the song) of the three files you created in question 3.  Does the reported polarity match your understanding of the song's lyrics? Why or why not do you think that might be?  Answer the questions in either a comment in the code cell or a markdown cell under the code cell.

```python
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob

# Load spaCy model with TextBlob
nlp = spacy.load("en_core_web_sm")
nlp.add_pipe("spacytextblob")

def analyze_sentiment(filename):
    with open(filename, "r") as file:
        lyrics = file.read()
    doc = nlp(lyrics)
    polarity_score = doc._.polarity
    return polarity_score

# Test the function by printing the polarity scores of the three files created in question 3
if __name__ == "__main__":
    files = ["shape_of_you_lyrics.txt", "someone_like_you_lyrics.txt", "fix_you_lyrics.txt"]
    for file in files:
        song_name = file.split("_lyrics.txt")[0].replace("_", " ").title()
        polarity_score = analyze_sentiment(file)
        print(f"Polarity score of '{song_name}': {polarity_score}")
```

### Explanation:

The reported polarity scores may or may not match the understanding of the song's lyrics. This depends on various factors, including the complexity and ambiguity of the lyrics, the context of the song, and the individual interpretation.

For example, sentiment analysis algorithms may struggle with understanding sarcasm, irony, or figurative language present in lyrics. Additionally, the sentiment analysis model used may have limitations in capturing the nuances of emotional expression in music.

In some cases, the reported polarity scores may align well with the perceived sentiment of the lyrics, especially in songs with straightforward and unambiguous emotional content. However, in other cases, there may be discrepancies between the reported polarity scores and the intuitive understanding of the song's sentiment.

Ultimately, the reported polarity scores provide a quantitative assessment of sentiment, but they should be interpreted alongside the listener's subjective understanding and emotional response to the lyrics and music.
